## Building ML Model to prediction Teams position based on season's performance

Trying to build an ML model to predict the position of teams in the EPL based on the 2021/2022 Season. 

```python
import pandas as pd
```

```python
df = pd.read_csv('EPL Standings 2000-2022.csv')
```

```python
df.head()
```

```python
df[df.Season == '2021-22']
```

```python
data2022 = df[df.Season == '2021-22']
```

```python
data2022.drop(['Season','Qualification or relegation','Team','Pld'],axis = 1,inplace = True)
```

```python
data2022
```

```python
y = data2022.Pos.values.reshape(-1,1)
X = data2022.drop(['Pos','D','GD'],axis = 1).values
```

```python
X
```

```python
#building baseline model
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
```

```python
model1.fit(X,y)
```

```python
model1.score(X,y)
```

```python
model1.predict(X)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
model2 = LinearRegression()
```

```python
model2.fit(X_scaled,y)
```

```python
model2.score(X_scaled,y)
```

```python
model2.predict(X_scaled)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model3 = RandomForestRegressor()
```

```python
model3.fit(X_scaled,y)
```

```python
model3.score(X_scaled,y)
```

```python
model3.predict(X_scaled)
```

```python
from sklearn.linear_model import Lasso
```

```python
model4 = Lasso()
model4.fit(X_scaled,y)
```

```python
model4.score(X_scaled,y)
```

```python
model4.predict(X_scaled)
```

```python
from catboost import CatBoostRegressor
```

```python
ctb = CatBoostRegressor()
```

```python
ctb.fit(X_scaled,y,silent=True)
```

```python
ctb.predict(X_scaled)
```

```python
df
```

```python
#Test Sets
Testset1 = df[df.Season == '2020-21']
Testset1
```

```python
y_test1 = Testset1.Pos.values.reshape(-1,1)
X_test1 = Testset1.drop('Pos',axis = 1)
X_test1.drop(['Season','Qualification or relegation','Team','Pld','D','GD'],axis = 1,inplace = True)
```

```python
X_test1_scaled = scaler.fit_transform(X_test1)
```

```python
ctbpred2021 = ctb.predict(X_test1_scaled).reshape(-1)
ctbpred2021
```

```python
linearpred2021 = model2.predict(X_test1_scaled).reshape(-1)
linearpred2021
```

```python
Rfpred2021 = model3.predict(X_test1_scaled).reshape(-1)
Rfpred2021
```

```python
results = pd.DataFrame({'LinearPred': linearpred2021,'CatboostPred':ctbpred2021,'RfPred':Rfpred2021})
```

```python
results
```

```python
results['Linearappox'] = results.LinearPred.apply(lambda x: 1 if x<1 else 20 if x>20 else round(x))
```

```python
results
```

```python
results['Catboostapprox'] = results.CatboostPred.apply(lambda x: 1 if x<1 else round(x))
results
```

```python
results['Rfapprox'] = results.RfPred.apply(lambda x: 1 if x<1 else 20 if x>20 else round(x))
results
```

```python
data = pd.concat([Testset1.reset_index(),results.reset_index()],axis=1)
```

```python
data.drop(['index','Qualification or relegation','GF','GA','W','L','D','Pld'],axis=1,inplace = True)
```

```python
data.to_csv('EPL 2021 prediction.csv',index=False)
```
